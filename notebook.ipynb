# <font color= #6B6B6B> 3.1 Libros de ordenes </font>

<hr style="border:0.8in solid gray"> </hr>

Un libro de órdenes u Order Book por su nombre en inglés, es una lista de operaciones, ya sean electrónicas o manuales, que una bolsa utiliza para registrar el interés del mercado en un valor específico o instrumento financiero. 
Las acciones se cotizan normalmente en un libro de órdenes por volumen y por nivel de precios.
Se actualiza en el tiempo real y permite medir el "sentimiento" del mercado en un determinado momento.
